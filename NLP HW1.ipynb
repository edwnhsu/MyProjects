# Select only the records corresponding to failed inspection (see "results" column)

```python
# import data
import csv
import pandas as pd
df = pd.read_csv('Food_Inspections_20240108.csv')
df.head(10)
```

```python
df.shape
```

```python
# Select only the records where the inspection result is 'Fail'
failed_inspections = df[df['Results'] == 'Fail']

# Display the DataFrame with only failed inspections
failed_inspections.head(10)
```

```python
failed_inspections.shape
```

# Clean the data, making sure that there are no NaNs in "violations" column

```python
# Drop rows with NaN values in the "violations" column
df_cleaned = failed_inspections.dropna(subset=['Violations'])
df_cleaned.head(10)
```

```python
df_cleaned.shape
```

### "Violations" column lists the reasons for inspection failure. Those reasons are separated by "|".  Each reason consists of a regulation code, regulation description, and comments describing how the regulation was violated. 

# Using regular expression, parse "violations" column to select only regulation descriptions, no code or comments

```python
# Define a regular expression pattern to extract the regulation descriptions
regulation_description_pattern = r'\d+\.\s*([^|]+) -'

# Apply the regular expression and create a new column with only the regulation descriptions
df_cleaned['regulation_descriptions'] = df_cleaned['Violations'].str.findall(regulation_description_pattern)

# Display the DataFrame with the new column
df_cleaned.head(10)
```

```python
print(df_cleaned["regulation_descriptions"])
```

```python
df_cleaned.loc[331, 'regulation_descriptions']
```

# Count how many times each regulation description occurred in the table and visualize top-10 the most frequent regulation descriptions

```python
# Flatten the list of regulation descriptions
flat_descriptions = [item for sublist in df_cleaned['regulation_descriptions'] for item in sublist]

# Create a Series to count the occurrences of each regulation description
description_counts = pd.Series(flat_descriptions).value_counts()

# Visualize the top 10 most frequent regulation descriptions
top_10_descriptions = description_counts.head(10)
top_10_descriptions
```

```python
top_10_descriptions.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Most Frequent Regulation Descriptions')
plt.xlabel('Regulation Descriptions')
plt.ylabel('Count')
plt.show()
```

```python
# Group the data by name and address
unique_buiz = df_cleaned.groupby(["DBA Name", "Address", "License #"]).agg({
    'regulation_descriptions': lambda x: x.tolist(),
    'Violations': 'count'
}).reset_index()

# Rename the 'Violations' column to 'count'
unique_buiz = unique_buiz.rename(columns={'Violations': 'count'})

# Display the resulting DataFrame
print(unique_buiz)
```

```python
# Filter rows with count more than 1
repeat_offenders = unique_buiz[unique_buiz['count'] > 1]
repeat_offenders
```

### Food inspection data has 10+ years of history, do you see any changing trends in most common violations?

# Plot the results for top-5 most frequent violations over time

```python
df_cleaned['Inspection Date'] = pd.to_datetime(df_cleaned['Inspection Date'])
# Extract year from the inspection date
df_cleaned['Inspection Year'] = df_cleaned['Inspection Date'].dt.year
df_cleaned
```

```python
# Explode the 'regulation_descriptions' column to separate rows for each violation description
exploded_df = df_cleaned.explode('regulation_descriptions')

# Group the data by year and regulation description, then count occurrences
grouped_data = exploded_df.groupby(['Inspection Year', 'regulation_descriptions']).size().reset_index(name='count')

grouped_data.head(50)
```

```python
# Find the Top-5 most frequent violations for each year
top_5_per_year = grouped_data.groupby(['Inspection Year', 'regulation_descriptions'])['count'].sum() \
    .groupby('Inspection Year').nlargest(5).reset_index(level=0, drop=True).reset_index()

top_5_per_year.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot the results for the Top-5 most frequent violations for each year
plt.figure(figsize=(14, 8))
sns.lineplot(data=top_5_per_year, x='Inspection Year', y='count', hue='regulation_descriptions', marker='o', palette='tab10')
plt.title('Top-5 Most Frequent Violations Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Regulation Description', bbox_to_anchor=(1, 1))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Find the overall Top-5 most frequent violations
top_5_overall = grouped_data.groupby('regulation_descriptions')['count'].sum().nlargest(5).reset_index()

# Filter the data for the top 5 overall violations
filtered_data = grouped_data[grouped_data['regulation_descriptions'].isin(top_5_overall['regulation_descriptions'])]

# Create a seaborn lineplot for the trend
plt.figure(figsize=(12, 8))
sns.lineplot(x='Inspection Year', y='count', hue='regulation_descriptions', data=filtered_data, marker='o')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trend of Top 5 Violations Over Time')

# Show the plot
plt.show()
```
